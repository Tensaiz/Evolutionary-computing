```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
'''
Read the log file as pandas dataframe

Use generation as index and skip the latter parameter rows 
and eliminate white space from columns
'''
def parse_log_file(file_path):
    df = pd.read_csv(file_path, sep='\t', index_col='gen', skiprows=lambda x: x > 31)
    df.columns = df.columns.str.replace(' ', '')
    return df
```

```python
PATH_PREFIX = 'evoman_framework/specialist_A1_fixed/'
file_name = 'life_test01-10-2019_01-42-38_enemy_1.txt'

df = parse_log_file(PATH_PREFIX + file_name)
df
```

```python
PATH_PREFIX = 'evoman_framework/specialist_A1_fixed/'
files = os.listdir(PATH_PREFIX)
file_names = [file for file in files if 'life_test' in file]
```

```python
file_names_e1 = [file for file in file_names if 'enemy_1' in file]
file_names_e2 = [file for file in file_names if 'enemy_2' in file]
file_names_e3 = [file for file in file_names if 'enemy_3' in file]

files_enemies_map = {
    'files_e1': file_names_e1,
    'files_e2': file_names_e2,
    'files_e3': file_names_e3
}
```

```python
grouped_row_enemy_map = {
    'grouped_row_e1': None,
    'grouped_row_e2': None,
    'grouped_row_e3': None,
}

for i in range(1, 4):
    df_concat = None
    # Concatenate all the results in one big dataframe (for one enemy)
    for f in files_enemies_map['files_e' + str(i)]:
        df = parse_log_file(PATH_PREFIX + f)
        if df_concat is None:
            df_concat = df
            continue
        df_concat = pd.concat((df_concat, df))
    # Group them by row so we can keep the structure and do statistical analysis
    grouped_row_index = df_concat.groupby(df_concat.index)
    grouped_row_enemy_map['grouped_row_e' + str(i)] = grouped_row_index
```

Averaged mean over specified runs (independent experiments)

and

Averaged std over specified runs (independent experiments)

```python
means_enemies_map = {
    'df_means_e1': None,
    'df_means_e2': None,
    'df_means_e3': None
}

std_enemies_map = {
    'df_std_e1': None,
    'df_std_e2': None,
    'df_std_e3': None
}

for i in range(1, 4):
    means_enemies_map['df_means_e' + str(i)] = grouped_row_enemy_map['grouped_row_e' + str(i)].mean()
    std_enemies_map['df_std_e' + str(i)] = grouped_row_enemy_map['grouped_row_e' + str(i)].std()
```

For each Task/algorithm, plot the average/standard deviation of the fitness and energy points (life) of the agents across the generations using a line-plot. Note that you need to calculate the average (over the 10 runs) of the average (over the populations).

```python
plt.subplot(2, 3, 1)
```

```python
# Average fitness + std

x = list(range(0, 31))

fig = plt.figure(figsize=(12, 3))
for i in range(1, 4):
    plt.subplot(1, 3, i)
    plt.errorbar(x, means_enemies_map['df_means_e' + str(i)]['avg'], yerr=std_enemies_map['df_std_e' + str(i)]['avg'], label="Mu + Lambda")
    plt.ylabel('Average fitness')
    plt.xlabel('Generation')
    plt.legend()
    plt.title('Enemy ' + str(i))
plt.show()

fig = plt.figure(figsize=(12, 3))
for i in range(1, 4):
    plt.subplot(1, 3, i)
    # Average life + std
    plt.errorbar(x, means_enemies_map['df_means_e' + str(i)]['life_avg'], yerr=std_enemies_map['df_std_e' + str(i)]['life_avg'], label="Mu + Lambda")
    plt.ylabel('Average life')
    plt.xlabel('Generation')
    plt.legend()
    plt.title('Enemy ' + str(i))
plt.show()
```

For each Task/algorithm/enemy, test your final best solution for each of the 10 independent runs, and present this result in a box-plot. Additionally, do a statistical test to verify if the differences in average are significant.

```python
best_fitnesses = {
    'best_fitness_a1_e1': [],
    'best_fitness_a1_e2': [],
    'best_fitness_a1_e3': [],

    'best_fitness_a2_e1': [],
    'best_fitness_a2_e2': [],
    'best_fitness_a2_e3': []
}

for e in range(1, 4):
    for f in file_names:
        df = parse_log_file(PATH_PREFIX + f + str(e) + '.txt')
        best_fitnesses['best_fitness_a1_e' + str(e)].append(max(df['max']))

best_fitnesses
```

```python
fig = plt.figure(figsize=(12, 3))
for i in range(1, 4):
    plt.subplot(1, 3, i)
    
    best_f_list = best_fitnesses['best_fitness_a1_e' + str(i)]
    
    # Boxplot of x-run averages of average population fitness
    plt.boxplot(best_f_list)
    # Plot arithmetic mean
    plt.scatter(1.15, np.mean(best_f_list))
    # Hide x-axis ticks
    plt.xticks([1], ['Mu + Lambda'])
    if i == 1:
        plt.ylabel('Best fitness')
    
    plt.title('Best solutions for enemy ' + str(i))
plt.show()
```
